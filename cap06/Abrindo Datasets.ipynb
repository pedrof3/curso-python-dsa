### Dividindo um arquivo em colunas

```python
file = open("arquivos/Current_Employee_Names__Salaries__and_Position_Titles_20240107.csv", "r")
data = file.read()
rows = data.split('\n')
full_data = []
```

```python
for row in rows:
    split_row = row.split(",")
    full_data.append(split_row)
```

### Contando número de linhas do Arquivo

```python
file = open("arquivos/Current_Employee_Names__Salaries__and_Position_Titles_20240107.csv", "r")
data = file.read()
rows = data.split('\n')
```

```python
cont = 0
for row in rows:
    cont += 1

print(f"Número de linhas do arquivo: {cont}")
```
